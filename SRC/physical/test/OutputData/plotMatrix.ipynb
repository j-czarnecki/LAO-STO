```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hamiltonianName = "H_rashba_rotated"

matrixReal = np.loadtxt(f"{hamiltonianName}_real.dat")
matrixImag = np.loadtxt(f"{hamiltonianName}_imag.dat")
```

```python
plt.pcolormesh(matrixReal, cmap="seismic", edgecolors="black", linewidth=0.5)
plt.gca().invert_yaxis()
plt.colorbar(label="Value")
plt.show()
```

```python
plt.pcolormesh(matrixImag, cmap="seismic", edgecolors="black", linewidth=0.5)
plt.gca().invert_yaxis()
plt.colorbar(label="Value")
plt.show()
```
